```python
import numpy as np

def cost_function(w):
    # 定义成本函数为 w^2
    return (w-5)**2
```

```python
import matplotlib.pyplot as plt

# 定义权重值范围
w_range = np.arange(0, 10, 0.1)

# 计算每个权重值的成本
costs = [cost_function(w) for w in w_range]

# 绘制成本函数图形
plt.plot(w_range, costs)
plt.xlabel('Weight')
plt.ylabel('Error')
plt.title('Error Function E(w)')

# 在w=9的点处标记
plt.scatter(9, cost_function(9), color='red')

plt.scatter(5, cost_function(5), color='red')


plt.show()
```

```python
import matplotlib.pyplot as plt

# 定义权重值范围
w_range = np.arange(0, 10, 0.1)

# 计算每个权重值的成本
costs = [cost_function(w) for w in w_range]

# 计算w=9处的梯度
gradient = 2 * 9

# 计算切线的截距
intercept = cost_function(9) - gradient * 9

# 绘制成本函数图形和切线
plt.plot(w_range, costs)
plt.plot(w_range, gradient * w_range + intercept, '--', color='red')
plt.xlabel('Weight')
plt.ylabel('Cost')
plt.title('Cost function E(w)')

# 在w=9的点处标记
plt.scatter(9, cost_function(9), color='red')

plt.show()
```
