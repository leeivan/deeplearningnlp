```python
import matplotlib.pyplot as plt
import numpy as np

# 定义x和y的范围
x = np.linspace(-10, 10, 100)
y = x

# 绘制坐标系和线性函数
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)

# 显示图形
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 定义x和y的范围
x = np.linspace(-5, 5, 1000)
y = np.zeros_like(x)

# 计算二元阶跃函数的值
y[x >= 0] = 1

# 绘制坐标系和二元阶跃函数
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)

# 显示图形
plt.show()
```

```python
import numpy as np

def cost_function(w):
    # 定义成本函数为 w^2
    return (w-5)**2
```

```python
import matplotlib.pyplot as plt

# 定义权重值范围
w_range = np.arange(0, 10, 0.1)

# 计算每个权重值的成本
costs = [cost_function(w) for w in w_range]

# 绘制成本函数图形
plt.plot(w_range, costs)
plt.xlabel('Weight')
plt.ylabel('Error')
plt.title('Error function E(w)')

# 在w=9的点处标记
plt.scatter(9, cost_function(9), color='red')

plt.show()
```
